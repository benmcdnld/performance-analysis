```python
import numpy as np
import pandas as pd
import drawdown
```

*This section pulls return data from csv*

```python
# Input csv file name located on Desktop that contains returns
file = 'Ivy.csv'

# Input workbook name results will be saved to
wbname = "Ivy Drawdowns"

csvfile = r'C:\Users\bmcdonald\Desktop\\' + file
df = pd.read_csv(csvfile, index_col=0)
date = df.index
returns = df.iloc[:,0]
```

*This section creates dataframe of top 20 drawdowns*

```python
dd = drawdown.maxdd(date, returns)
dd
```

*This section writes the results to Excel*

```python
writer = pd.ExcelWriter(wbname, engine='xlsxwriter')

dd.to_excel(wbname, index=False)

writer.save()
```
